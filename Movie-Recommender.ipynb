# Method 1. Using IMDB Weighted Rating (For top rated movie recommendation)

This formula provides a true 'Bayesian estimate', which takes into account the number of votes each title has received, minimum votes required to be on the list, and the mean vote for all titles:

weighted rating (WR) = (v ÷ (v+m)) × R + (m ÷ (v+m)) × C

Where:

R = average for the movie (mean) = (rating)

v = number of votes for the movie = (votes)

m = minimum votes required to be listed in the Top Rated list

C = the mean vote across the whole report


Source :- IMDB

```python
C = data['vote_average'].mean()
C
```

This means mean rating for all the movies is 6 out of 10.
Now we will set our threshold to 90%, i.e, minimum number of votes required to be listed in the Top Rated List should be atleast 90%. Or in other words we have to leave 10%.

```python
#There are 2 ways we can calculate 'm'

#1. using quantile() 
# m = data['vote_count'].quantile(0.9)

#2. Using formula -> int(q(n+1))
m = int(0.1*(len(data)+1))
m
```

```python
def weighted_rating(df,m, C):
    v = df['vote_count']
    R = df['vote_average']
    # Calculation based on the IMDB formula
    return (v/(v+m) * R) + (m/(m+v) * C)
```

### Filtering the movies that can be featured in chart

```python
feat_movies = data.copy().loc[data['vote_count'] >= m]
feat_movies.head()
```

```python
# Adding a new column 'weighted_rating', which will contain weighted_ratig of all featured movies
feat_movies['weighted_rating'] = feat_movies.apply(weighted_rating,args=(m,C), axis=1)
```

```python
#Sorting movies based on weighted_rating
feat_movies = feat_movies.sort_values('weighted_rating', ascending=False)

#Top 25 movies 
feat_movies[['original_title', 'vote_count', 'vote_average', 'weighted_rating']].head(25)
```

This will recommend similar movies based on the given movie.

Features that will be considered while finding similar movies are :- keywords, genres, cast, crew

###  Step (b). Extracting our features in structured format from the dataset

### * Our data is present in stringified list format , we will convert it into usable format

**literal_eval** - ast.literal_eval raises an exception if the input isn't a valid Python datatype, so the code won't be executed if it's not.
It will convert "[1,2]" into [1,2]

```python
from ast import literal_eval

features = ['cast', 'crew', 'keywords', 'genres']
for feature in features:
    data[feature] = data[feature].apply(literal_eval)
```

### * Getting values like director, actor names and and keywords, genres

```python
def get_director(crew):
    for i in crew:
        if i['job'] == 'Director':
            return i['name']
    return np.nan

#Getting director
data['director'] = data['crew'].apply(get_director)
```

```python
def get_imp_data_from_feat(feat):    
    
    #check if data is not malformed/missing
    if isinstance(feat, list):
        names = [i['name'] for i in feat]
        
        #we will use only top 4 infos.
        if len(names) > 4:
            names = names[:4]
        return names

    #Return empty list in case of missing/malformed data
    return []


features = ['cast', 'keywords', 'genres']
for feature in features:
    data[feature] = data[feature].apply(get_imp_data_from_feat)
```

```python
#Preview the modified dataset
data[['original_title', 'cast', 'director', 'keywords', 'genres']].head()
```

### Step (c). We will use a Natural language Processing technique called as Feature Extraction using Count Vectorizer, so that our model will understand the relation between all the data that we obtained.

### * Clean the available data(converting everything in lower case to avoid ambiguity and removing spaces)

```python
#Cleaning the data

def clean_data(x):
    
    #when our data is contained in list
    if isinstance(x, list):
        return [str.lower(i.replace(" ", "")) for i in x]
    else:
        #when our data is string
        if isinstance(x, str):
            return str.lower(x.replace(" ", ""))
        else:
            return ''
        
features = ['cast', 'keywords', 'director', 'genres']

for feature in features:
    data[feature] = data[feature].apply(clean_data)
```

### * Now we'll combine all available data we have into a single string format for each movies

```python
def combining_metadata(x):
    keywords = ' '.join(x['keywords'])
    cast = ' '.join(x['cast'])
    director = ' '.join(x['director'])
    genres = ' '.join(x['genres'])
    
    
    return keywords + " " + cast + " " + director + " " + genres
data['soup'] = data.apply(combining_metadata, axis=1)
```

### * Implementing Count Vectorizer

We will also remove stop_words from our data.

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vect = CountVectorizer(stop_words='english')
count_matrix = count_vect.fit_transform(data['soup'])
```

### * Calculating the Cosine Similarity matrix based on the count_matrix

scikit-learn already provides pairwise metrics (a.k.a. kernels in machine learning parlance) that work for both dense and sparse representations of vector collections. In this case we need a dot product that is also known as the linear kernel

Source :- https://stackoverflow.com/questions/12118720/python-tf-idf-cosine-to-find-document-similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(count_matrix, count_matrix)
```

```python
def recommender(title, similarity_variable, num):
    
#     global sim_scores, recom_movie #will be used in plotting graph
    # Get the index of the movie that matches the title
    index = title2index[title]

    # Get the pairwsie similarity scores of all movies with that movie
    sim_scores = list(enumerate(similarity_variable[index]))

    # Sort the movies based on the similarity scores
    
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Getting the scores of the 'num' most similar movies
    #At 0th index, te same movie will exist. So, we start from 1.
    sim_scores = sim_scores[1:num+1]

    # Get the movie indices
    movie_indices = [i[0] for i in sim_scores]

    # Return the top 'num' most similar movies
    recom_movie = data['original_title'].iloc[movie_indices]
    return recom_movie
```
